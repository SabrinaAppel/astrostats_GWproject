```python
import linmix
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,100,50)
y = 0.5*x+4
xerr = np.ones(len(x))
yerr=np.ones(len(y))

lx = linmix.LinMix(x,y,xsig=xerr,ysig=yerr)
lx.run_mcmc(silent=True)

zeropoints = np.median(lx.chain[:]['alpha'])
slopes = np.median(lx.chain[:]['beta'])
scatters = np.sqrt(np.median(lx.chain[:]['sigsqr']))
```

```python
len(lx.chain['alpha'])
```

```python
plt.plot(x+xerr,y+yerr)
plt.plot(x,x*slopes+zeropoints)
for i in range(100):
    plt.plot(x,x*lx.chain[i]['beta']+lx.chain[i]['alpha'])
```
